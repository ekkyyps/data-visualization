#### Code

```python
# Drop table
df_cor = dataset.drop(columns='Outcome')
# Correlation between Variables without Outcome because it use to be Target
cor = df_cor.corr(method='pearson')
```

```python
plt.figure(figsize=(9,7))
sns.heatmap(cor, 
            xticklabels=cor.columns,
            yticklabels=cor.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
plt.title('(i) Correlation Between Variables')
plt.show()
```

```python
# Correlation between Variable and Target
df_cor2 = dataset.copy()

figure = plt.subplots(figsize = (8,7))
plt.title('(ii) Correlation between Variables and Target')
sns.heatmap(df_cor2.corr()[['Outcome']], annot = True, fmt='.1g',vmin=-1, vmax=1, center= 0)
```

Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.

#### Code

```python
# Load the dataset
dataset_hang = pd.read_csv("Hang_Seng.csv")
dataset_nasd = pd.read_csv("Nasdaq.csv")
dataset_nikk = pd.read_csv("Nikkei.csv")
```

Preprocessing

```python
# Checking null dataset Han_Seng
dataset_hang.isnull().sum()
# Handling with drop the null
dataset_hang.dropna(inplace = True, axis = 0)
```

```python
# Checking null dataset Nasdaq
dataset_nasd.isnull().sum()
# Handling with drop the null
dataset_nasd.dropna(inplace = True, axis = 0)
```

```python
# Checking null dataset Nikkei
dataset_nikk.isnull().sum()
# Handling with drop the null
dataset_nikk.dropna(inplace = True, axis = 0)
```

```python
df_hang = dataset_hang.round(2)
df_nssd = dataset_nasd.round(2)
df_nikk = dataset_nikk.round(2)
```

```python
df_hang.head(3)
```

Calculate Daily Percentage

```python
df_hang["Day_Perc_Change"] = df_hang["Adj Close"].pct_change()*100
df_hang.head(3)
```

```python
df_nssd["Day_Perc_Change"] = df_nssd["Adj Close"].pct_change()*100
df_nssd.head(3)
```

```python
df_nikk["Day_Perc_Change"] = df_nikk["Adj Close"].pct_change()*100
df_nikk.head(3)
```

```python
df_hang.iloc[-90:].describe().astype(int)
```

Trend Analysis

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
```

```python
df_hang["Trend"]= df_hang["Day_Perc_Change"].apply(lambda x:trend(x))
# df_hang.head()
df_nssd["Trend"]= df_nssd["Day_Perc_Change"].apply(lambda x:trend(x))
# df_nssd.head()
df_nikk["Trend"]= df_nikk["Day_Perc_Change"].apply(lambda x:trend(x))
# df_nikk.head()
```

Set Date

```python
from datetime import datetime
```

```python
start = '2019-01-01'
end = '2019-12-31'
df_hang = df_hang.loc[(df_hang['Date'] >= start) & (df_hang['Date'] <= end)]
df_nssd = df_nssd.loc[(df_nssd['Date'] >= start) & (df_nssd['Date'] <= end)]
df_nikk = df_nikk.loc[(df_nikk['Date'] >= start) & (df_nikk['Date'] <= end)]
```

Pie Chart Han_Seng

```python
pie_data = df_hang.groupby(['Trend'])
pie_label = sorted([i for i in df_hang.loc[:, 'Trend'].unique()])
plt.pie(pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 1)
plt.title("Trend Analysis of Han_Seng", bbox={'facecolor':'0.8', 'pad':5})

plt.show()
```

Pie Chart Nasdaq

```python
pie_data_nssd = df_nssd.groupby(['Trend'])
pie_label_nssd = sorted([i for i in df_nssd.loc[:, 'Trend'].unique()])
plt.pie(pie_data_nssd['Trend'].count(), labels = pie_label_nssd, 
        autopct = '%1.1f%%', radius = 1)
plt.title("Trend Analysis of Nasdaq", bbox={'facecolor':'0.8', 'pad':5})

plt.show()
```

```python
pie_data_nikk = df_nikk.groupby(['Trend'])
pie_label_nikk = sorted([i for i in df_nikk.loc[:, 'Trend'].unique()])
plt.pie(pie_data_nikk['Trend'].count(), labels = pie_label_nikk, 
        autopct = '%1.1f%%', radius = 1)
plt.title("Trend Analysis of Nikkei", bbox={'facecolor':'0.8', 'pad':5})

plt.show()
```

#### Penjelasan 

Berdasarkan hasil visualisasi dari tren "Slight or No Change" untuk saham Han_Seng selama transaksi dari periode Januari 2019 hingga Desember 2019 didapatkan sebanyak 48,4%. Sedangkan untuk saham Nasdaq sebanyak 46% dan saham Nikkei sebesar 52,7%. Hal ini berarti dari ketiga data tersebut mengalami perubahan harga saham yang setiap harinya(hari ini dengan kemarin) tidak jauh berbeda sehingga potensi dari ketiganya memiliki keuntungan yang kecil. 

_"**Persentase** tren **Slight ot No Change** berbanding terbalik dengan **Potensi Keuntungan** yang didapatkan"_

Dapat disimpulkan bahwa Nikkei memiliki keuntungan yang terkecil karena persentase untuk tren "Slight or No Change" memiliki persentase paling besar diantara saham Han_Seng dan saham Nasdaq.

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
df_combination= pd.concat([df_hang["Adj Close"], df_nssd["Adj Close"], df_nikk["Adj Close"]], axis=1)
df_combination.columns = ["Hang_Seng", "Nasdaq", "Nikkei"]

df_combination.dropna(inplace = True, axis = 0)
df_percent_combined = df_combination.pct_change()*100
df_percent_combined.dropna(inplace = True, how = "any", axis = 0)

df_percent_combined
```

```python
volatility = df_percent_combined.rolling(7).std()*np.sqrt(7)
volatility.plot(figsize = (15,7))

plt.title('Volatility of Han_Seng, Nasdaq, and Nikkei')
plt.show()
```

#### Penjelasan 

Dapat dilihat bahwa Han_Seng cenderung volatilitasnya stabil daripada Nasdaq. Nasdaq cenderung memiliki volatilitas yang tinggi daripada Han_Seng dan Nikkei. Hal ini akan menarik banyak pedagang dan investor karena mereka mencari investasi volatilitas yang lebih tinggi untuk mendapatkan keuntungan yang lebih tinggi. Jika suatu saham tidak bergerak, tidak hanya memiliki volatilitas yang rendah, tetapi juga potensi keuntungan yang rendah. Di sisi lain, saham atau sekuritas lain dengan tingkat volatilitas yang sangat tinggi dapat memiliki potensi keuntungan yang luar biasa, tetapi risikonya juga sama tingginya. Sedangkan, jika pemula lebih baik mencari yang stabil seperti milik Han_Seng karena risiko juga tidak terlalu tinggi.
